## **Penjelasan Dataset**

Dataset 'World Happiness Report' yang digunakan bisa diperoleh di folder data.

Dataset-dataset berikut ini menunjukkan score happines dunia. Score happiness menggunakan data dari GallUp world poll. Data feature yang menjadi kata kunci di report ini adalah:

*  **Life Ladder** >> Ibarat sebuah tangga, pijakan 0 adalah dasar dan 10 adalah atas. Tangga atas menunjukkan  kehidupan terbaik  dan dasar tangga menunjukkan kehidupan terburuk

*  **Social support**  >> dukungan sosial seperti keluarga, teman, tetangga dll

*   **Freedom to make life choices** >> Peluang tiap individu untuk melakukan aksi yang dipilih
*   **Generosity** >> Kemurahan hati
*  **Perceptions of corruption**  >> Korupsi

# Soal 2: Happiness report negara Indonesia dari tahun 2006 - 2021

Output yang diharapkan adalah data negara Indonesia dari penggabungan dataset happiness report dan dataset happiness report 2021. Lakukan perintah-perintah berikut:


1.   Lakukan filtering dataset happiness report dari kolom ['Country name']== 'Indonesia'
     - Drop kolom 'Positive affect' dan 'Negative affect'

2.   Lakukan filtering dataset dari variabel no 2 hasil dropping beberapa kolom
     - Tambahkan kolom ['year']=2021 pada dataset dari variabel no 2 hasil dropping beberapa kolom
     - Filtering dataset tersebut menggunakan kolom ['Country name']== 'Indonesia'
     - Rename beberapa kolom :
                  'Ladder score':'Life Ladder',
                  'Logged GDP per capita':'Log GDP per capita',
                  'Healthy life expectancy':'Healthy life expectancy at birth'

     - Drop kolom 'Dystopia + residual' dan 'Regional indicator'

3. Gabungkan data dari hasil poin 1 dan poin 2
    - Rename kolom {'year':'Year'}


```python
# Filtering dataset 1

df1.drop(columns = ['Positive affect','Negative affect'], inplace=True)
df1_indo = df1[(df1['Country name'] == 'Indonesia')].reset_index(drop = True)
```

```python
#Filtering dataset 2

df2.drop(columns=['Regional indicator','Dystopia + residual'],inplace=True)
df2.rename(columns={'Ladder score' :'Life Ladder',
                    'Logged GDP per capita': 'Log GDP per capita',
                    'Healthy life expectancy':'Healthy life expectancy at birth'},inplace=True)
df2_indo = df2[(df2['Country name'] == 'Indonesia')].assign(year=2021).reset_index(drop = True)
```

```python
#Penggabungan dataset

column_title = ['Country name','Year','Life Ladder','Log GDP per capita',
                'Social support','Healthy life expectancy at birth',
                'Freedom to make life choices','Generosity',
                'Perceptions of corruption']

df3 = pd.concat([df1_indo,df2_indo],sort=True).reset_index(drop = True)
df3.rename(columns={'year' : 'Year'}, inplace=True)

#Index ulang kolom
df = pd.DataFrame(df3, columns = column_title)
df
```

# Soal 3: Visualisasi lineplot Life Ladder negara Indonesia


```python
plt.figure(figsize = (14,6))
sns.set_style('whitegrid')
sns.lineplot(data = df, 
            x = 'Year', 
            y = 'Life Ladder', 
            markers = True, 
            marker = 'o', 
            markersize = 8)
sns.despine(left = True) #Garis batas diluar grafik
plt.title("Indonesia's ladder score over the years", 
          fontweight = 'bold', 
          fontsize = 15)
plt.show()
```

# Soal 4: Visualisasi Scatter plot

Buatlah visualisasi scatter plot (perpaduan seaborn dan matplotlib) dari df2 antara GDP per capita (Log GDP per capita) dan  skor korupsi (Perceptions of corruption).

Gunakan kolom skor kebebasan beraksi ('Freedom to make life choices').

Gunakan kolom skor kesehatan ('Healthy life expectancy') sebagai warna point, berikan cmap 'plasma'.

Gunakan regplot pada seaborn (sns.regplot) dengan argumen scatter=False dan color = 'black'

```python
fig, axes = plt.subplots(figsize=(14,10))

datapoints = axes.scatter(x=df2['Log GDP per capita'],
                        y=df2['Perceptions of corruption'],
                        s=df2['Freedom to make life choices']*900,
                        c=df2['Healthy life expectancy at birth'], cmap='plasma', alpha=0.5)

plt.colorbar(datapoints)

sns.regplot(data=df2,
            x='Log GDP per capita',
            y='Perceptions of corruption',
            color='k',
            scatter=False,
            ax=axes)

axes.set_title('Corruption vs GDP percapita', weight='bold')

plt.show()
```
