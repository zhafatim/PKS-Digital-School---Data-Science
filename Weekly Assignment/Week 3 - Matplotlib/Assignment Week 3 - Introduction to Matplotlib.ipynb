# Introduction to Matplotlib

#### Component Figure & Axis

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
#Explicit

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x,y)
plt.show()
```

```python
#Implicit

plt.plot(x,y)
plt.show()
```

#### Memasukkan multiple data ke dalam satu axis

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
```

```python
#plot data
fig,ax = plt.subplots()
ax.plot(x,y)
ax.plot(x,y2,'g')
ax.plot(x,y3,'r')
plt.style.use('seaborn')
plt.show()
```

# Customization of Visualization

Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
fig = plt.figure(figsize = (10,6))
ax = fig.add_subplot()

ax.plot(x,cos,color = 'r',label = 'Cosx')
ax.plot(x,cos3,color = 'g',linestyle='--', label = '3 Cosx')
ax.plot(x,sin,color = 'b',linestyle='-.',label='Sin(x-90)')
ax.plot(x,sin2,color = 'y',linestyle = 'none',label='2 Sinx',marker = 'x')

ax.set_xlabel('X-Axis Label',fontsize = 'small')
ax.set_ylabel('Y-Axis Label',fontsize = 'small')
ax.set_title('Grafik Cos Sin',fontsize = 'small')
ax.legend(loc = 'center',fontsize = 'small')

plt.show()
```

# Multiple Subplots

Buat plot dengan data dibawah ini

#### No 1

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2*-np.pi, 2*np.pi, 200)

tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig,ax = plt.subplots(nrows=1,ncols=3, figsize = (16,8),sharey=True)
ax = ax.flatten()

ax[0].plot(x,tan,color = 'b',label='tanX')
ax[1].plot(x,sin,color='r',label='sinX')
ax[2].plot(x,cos,color='g',label='cosX')

ax[0].set_title('Grafik tanx',fontsize='small')
ax[1].set_title('Grafik sinx',fontsize='small')
ax[2].set_title('Grafik cosx',fontsize='small')

fig.legend(loc='upper center',fontsize='small')

plt.show()
```

#### No 2

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
fig, ax = plt.subplots(figsize=(14,9))

ax1 = ax.inset_axes((-0.1, 0.9, 0.3, 0.3))
ax2 = ax.inset_axes((-0.09, -0.1, 0.3, 0.3))
ax3 = ax.inset_axes((0.91, -0.1, 0.3, 0.3))
ax4 = ax.inset_axes((0.9, 0.9, 0.3, 0.3))

ax.set_title('Grafik CosX')
ax1.set_title('Grafik CosX^5')
ax2.set_title('Grafik CosX^2')
ax3.set_title('Grafik CosX^3')
ax4.set_title('Grafik CosX^4')

ax.plot(x,y,'b')
ax1.plot(x,y5,'y',x,y,'b')
ax2.plot(x,y2,'r',x,y,'b')
ax3.plot(x,y3,'g',x,y,'b')
ax4.plot(x,y4,'k',x,y,'b')

plt.show()
```

# Quantitive Comparison

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data/vgsales.csv')
df.head()
```

#### Preparing Data

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
sales_mean = df.groupby('Genre')[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].mean()
sales_mean
```

#### Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
sales_tr = sales_mean.transpose()

fig, ax = plt.subplots(figsize=(16,8))

x = sales_tr.index

sales_tr.plot(kind='bar', ax=ax, zorder=2, width=0.8)

ax.set_title('Mean Sales Video Game By Genre', size=12)
ax.set_xlabel('Region Sales', size='small')
ax.set_ylabel('Mean Sales', size='small')

ax.set_xticks(ticks=range(len(x)), labels=sales_tr.index, size='small',
              rotation='horizontal')

plt.legend(loc=(0.95,0.7),fontsize='small')
plt.style.use('ggplot')
plt.show()
```

###### Insight

Di wilayah NA, video game dengan genre platform memiliki rata-rata tertinggi penjualan. Sedangkan di wilayah EU dan wilayah lainnya, genre video game dengan rata-rata penjualan tertinggi adalah shooter. Sedangkan genre role-playing memiliki rata-rata penjualan tertinggi di wilayah JP.

####  Perbandingan Kuantitatif Barplot - Stack Barplot


1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
sales_tr = sales_mean.transpose()

fig, ax = plt.subplots(figsize=(14,8))

x = sales_tr.index

sales_tr.plot(kind='bar', stacked = True, ax=ax, zorder=2, width=0.8)

ax.set_title('Mean Sales Video Game By Genre', size=12)
ax.set_xlabel('Region Sales', size='small')
ax.set_ylabel('Mean Sales', size='small')

ax.set_xticks(ticks=range(len(x)), labels=sales_tr.index, size='small',
              rotation='horizontal')

plt.legend(loc=(0.95,0.7), fontsize='small')
plt.style.use('ggplot')
plt.show()
```

###### Insight

Berdasarkan grafik terlihat bahwa wilayah NA memiliki rata-rata penjualan video game tertinggi, diikuti wilayah EU, wilayah JP, dan terakhir wilayah lainnya. 

Terlihat dari ketebalan bar, untuk wilayah NA, penjualan tertinggi adalah video game dengan genre Platform. Untuk wilayah JP, video game genre Role-Playing memiliki penjualan tertinggi. Sedangkan untuk wilayah EU dan wilayah lainnya, video game genre shooter memiliki penjualan tertinggi.

# Seaborn

```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset('diamonds')
df.head()
```

#### Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan.

```python
sns.set({'figure.figsize':(13,13)})
palette = "ch:r=-.2,d=.3_r"
 
#code here

clarity_order = ['IF','VVS1','VVS2','VS1','VS2','SI1','SI2','I1']
sns.scatterplot(data = df,
                x = 'carat',
                y = 'price', 
                hue = 'clarity',
                size = 'depth',
                sizes = (1,8), 
                hue_order = clarity_order,
                palette = "ch:r=-.2,d=.3_r", 
                linewidth = 0)

plt.show()
```

#### Catplot

Buatlah sebuah Categorical plot dengan kind = 'boxplot'


```python
sns.set_style('ticks')
sns.catplot(data = df,
            x = 'clarity',
            y = 'price', 
            col = 'cut',
            kind = 'box', 
            showfliers = False)
plt.show()
```

#### Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price serta markers =  ['p', 'o', 's', 'D', 'X'] 

```python
df_ran = df.sample(n=300, random_state=123)

sns.set_style('darkgrid')
sns.pairplot(data = df_ran,
             vars = ['carat','depth','table','price'],
             corner = True,
             hue = 'cut', 
             markers = ['p','o','s','D','X'])
plt.show()
```

#### Joint Plot

Gunakan data yang sama dengan soal 3 untuk menghasilkan dua visualisasi joint plot antara carat dan price.

```python
with sns.axes_style('white'):
    sns.jointplot(data = df_ran, 
              x = 'carat', 
              y = 'price',
              kind = 'kde', 
              fill = False)
plt.show()
```

```python
sns.set_style('darkgrid')
sns.jointplot(data = df_ran, 
              x = 'carat', 
              y = 'price',
              xlim = (0.25,2.25),
              kind = 'reg')

plt.show()
```
