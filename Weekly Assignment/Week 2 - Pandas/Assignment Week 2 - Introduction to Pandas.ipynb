# Accessing & Indexing Dataframe

```python
import pandas as pd
import random
```

```python
pelajaran = ['Matematika', 'Fisika', 'Biologi', 'Kimia', 'Geologi', 'Ekonomi']
kelas = ['12A', '12B', '12C', '12D']

hasil_nilai = {
    'siswa_id' : range(1, 21),
    'pelajaran' : [random.choice(pelajaran) for i in range(20)],
    'nilai' : [random.choice(range(0, 100)) for i in range(20)],
    'jenis-kelamin' : [random.choice(['laki-laki', 'perempuan']) for i in range(20)],
    'kelas' : [random.choice(kelas) for i in range(20)]
}

df = pd.DataFrame(hasil_nilai)
df.head()
```

####  Indexing

- Lakukan indexing terhadap dataframe untuk mengakses column nilai dan kelas saja
- Lakukan indexing terhadap dataframe untuk mengakses row 5 sampai 10 beserta column pelajaran dan nilai

```python
display(df[['nilai', 'kelas']])
```

```python
df.loc[4:9,'pelajaran':'nilai']
```

#### Filter DataFrame Dengan Boolean

- Tampilkan semua data siswa pada mata pelajaran Ekonomi dan Fisika
- Tampilkan semua data siswa yang mempunyai nilai lebih dari 50 di mata pelajaran Kimia 
- Tampilkan semua data yang mempunyai nilai lebih dari 70, berjnis kelamin perempuan dan di kelas 12C

```python
#Data siswa pada mata pelajaran Ekonomi dan Fisika
df[(df.pelajaran == 'Ekonomi') | (df.pelajaran == 'Fisika')]
```

```python
#Menampilkan data siswa yang mempunyai nilai lebih dari 50 di mapel Kimia
df[(df.pelajaran == 'Kimia') & (df.nilai > 50)]
```

```python
#Menampilkan data siswa yang mempunyai nilai lebih dari 70, perempuan, dan kelas 12C

df[(df.nilai>70) & (df['jenis-kelamin'] == 'perempuan') & (df.kelas == '12C')]
```

#### Data Transformation

Lakukan data transformasi terhadap column 'nilai' dengan fungsi transformasi, berikut rumus dari normalisasi

![alt text](https://drive.google.com/uc?id=1wed-EbRSTA0JP1ceDp6lcKBqH2yJ4aa3)

```python
df['nilai'] = df[['nilai']].apply(lambda nilai : (nilai-nilai.min())/(nilai.max()-nilai.min()))
df.nilai
```

# DataFrame Re-Shapping

#### Reshapping Data

```python
import pandas as pd
import random
```

```python
data = {
    'Tipe Ujian' : ['UTS1', 'UAS1', 'UTS2', 'UAS2']*3,
    'Mata Pelajaran' : ['Fisika']*4 + ['Matematika']*4+['Statistika']*4,
    'Rata2 Waktu Belajar (menit)' : [random.choice(range(200, 800)) for i in range(12)],
    'Rata2 Nilai Ujian' : [random.choice(range(40, 100)) for i in range(12)] 
}
df = pd.DataFrame(data)
df
```

```python
#Pivot Data

nilai_pivot = df.pivot_table(index = 'Tipe Ujian', columns = 'Mata Pelajaran', 
                             values = ['Rata2 Nilai Ujian','Rata2 Waktu Belajar (menit)'], 
                             aggfunc = 'mean')
nilai_pivot
```

# Data Grouping

```python
import pandas as pd
```

```python
df = pd.read_csv('data/diamonds.csv')
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.head()
```

Columns Metadata:

carat: berat diamond

cut: kualitas diamond

color: warna diamond

clarity: level kemurnian diamonds

depth: tinggi dari diamond tabel

table: lebar  dari diamond tabel

price: harga diamond ($US)

x: panjang

y: lebar

z: kedalaaman

#### Process Grouping

```python
#Grup data berdasarkan cut (kualitas) dan melakukan agregasi fungsi rata-rata

df_cut = df.groupby('cut').mean()
df_cut
```

####  Grouping Some Column based on Multiple Criteria

Lakukan groupby pada data berdasarkan kriteria cut (kualitas) dan color (warna) secara berurutan, dengan aggregasi pada column carat, table, dan price saja. dengan fungsi agregasi rata-rata.

```python
df_cc = df.groupby(['cut','color'])[['carat','table','price']].agg('mean')
df_cc
```

#### Group by Multiple Aggregate based on Multiple Criteria

Lakukan groupby pada data berdasarkan kriteria cut(kualitas) dan color (warna) secara berurutan kemudian aggregasi objek group by nya dengan fungsi rata2 dan median.

```python
df_multiple = df.groupby(['cut','color'])[['carat','depth','table','price','x','y','z']].agg(['mean','median'])
df_multiple
```

#### Group By With Different Aggregation for specific Column

Lakukan group by pada data berdasarkan criteria cut(kualitas), kemudian aggreagate dengan fungsi max pada column price dan fungsi min pada column carat.

```python
df_spec = df.groupby('cut')[['price','carat']].agg({'price':'max','carat':'min'})
df_spec
```

#### Group By With Custom Function

Lakukan groupby pada data berdasarkan criteria cut(kualitas), kemudian lakukan aggregasi dengan fungsi standar deviasi pada column carat dan fungsi rentang nilai antara max dan min pada column price.

```python
import numpy as np

df_func = df.groupby('cut')[['price','carat']].agg({'price':np.ptp,'carat':'std'})
df_func
```

# Data Merging 

#### Penggabungan Data 1

Pada soal ini gunakan data names1881.csv dan names1981.csv, data ini berisi nama bayi yang lahir pada tahun 1881 atau 1981 di Amerika.

```python
import pandas as pd
```

```python
#read data

df1 = pd.read_csv('data/names1881.csv', names = ['Name','Gender','Count'])
df2 = pd.read_csv('data/names1981.csv', names = ['Name','Gender','Count'])

# periksa dimensi dari kedua data
print("Jumlah Dimensi data 'names1881.csv' :",df1.shape)
print("Jumlah Baris data 'names1981.csv' :",df2.shape)

# gabungkan data
df = pd.concat([df1,df2])
display(df)

# periksa dimensi data hasil gabungan
print("Jumlah dimensi data setelah penggabungan : ",df.shape)
```

#### Penggabungan Data 2

data1 = {
    'id' : [1, 2, 3, 4],

```python
# join data
df_outer_join = pd.merge (df1,df2,how = 'outer', on = 'id')
display(df_outer_join)

df_inner_join = pd.merge (df1,df2,how = 'inner', on = 'id')
display(df_inner_join)

df_left_join = pd.merge (df1,df2,how = 'left', on = 'id')
display(df_left_join)

df_right_join = pd.merge (df1,df2,how = 'right', on = 'id')
display(df_right_join)
```
